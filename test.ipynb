```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter):
    z = 0
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter

# Define the grid size and range
x_min, x_max = -2.5, 1.5
y_min, y_max = -2.0, 2.0
width, height = 800, 800

# Create a grid of complex numbers
x = np.linspace(x_min, x_max, width)
y = np.linspace(y_min, y_max, height)
X, Y = np.meshgrid(x, y)
C = X + 1j * Y

# Initialize arrays to store points
points = []

# Compute the Mandelbrot set
max_iter = 100
for i in range(C.shape[0]):
    for j in range(C.shape[1]):
        iter_count = mandelbrot(C[i, j], max_iter)
        if iter_count == max_iter:
            # This point is in the Mandelbrot set
            points.append((X[i, j], Y[i, j]))

# Convert to numpy array for plotting
points = np.array(points)

# Plotting
plt.figure(figsize=(10, 10))
plt.scatter(points[:, 0], points[:, 1], color='black', s=0.1, edgecolor='none')
plt.title('Mandelbrot Set (Points that Belong)')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)')
plt.show()
```
